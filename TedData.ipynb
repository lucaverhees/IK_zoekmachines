```python
import pandas as pd
import re
import datetime
```

```python
df1 = pd.DataFrame.from_csv('transcripts.csv')
df1 = df1.reset_index()
```

```python
df1['transcript'] = df1['transcript'].apply(lambda x: re.sub(r'\([^()]*\)', '', x))
```

```python
df1
```

```python
df2 = pd.DataFrame.from_csv('ted_main.csv')
df2 = df2.reset_index()
```

```python
df2['film_date'] = df2['film_date'].apply(lambda x: datetime.datetime.fromtimestamp(int(x)).strftime('%d/%m/%Y'))
df2['published_date'] = df2['published_date'].apply(lambda x: datetime.datetime.fromtimestamp(int(x)).strftime('%d/%m/%Y'))
```

```python
df2
```

```python
df3 = pd.merge(df1, df2, on = "url")
```

```python
df3
```

```python
df3.to_json('TedData.json')
```
